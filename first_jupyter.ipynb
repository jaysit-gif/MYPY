```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate data
x = np.linspace(-5, 5, 1000)  # Create 1000 points from -5 to 5
y = np.floor(x)  # Apply the greatest integer function (floor)

# Step 2: Create the step plot
plt.step(x, y, label='Greatest Integer Function ⌊x⌋', color='blue', linewidth=2)

# Step 3: Customize the plot
plt.title('Greatest Integer Function')
plt.xlabel('x')
plt.ylabel('⌊x⌋')
plt.grid(True, which='both', linestyle='--', alpha=0.7)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)

# Highlight the steps with markers at integer points
for i in range(-5, 6):
    plt.plot(i, i, 'ro', markersize=8, label='Integer points' if i == -5 else "")
    plt.plot(i, i, 'ro', fillstyle='none', markersize=8)  # Open circles for right endpoints

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
t = np.linspace(0, 2 * np.pi, 100)
x = np.cos(t)
y = np.sin(t)
plt.plot(x, y)
plt.axis('equal')
plt.title('Unit Circle')
plt.show()
```

```python
random_data = np.random.uniform(-5, 5, 100)  # 100 random numbers between -5 and 5
noise = np.random.normal(0, 0.1, 100)  # Gaussian noise

x = np.linspace(-5, 5, 100)
y = np.floor(x) + np.random.normal(0, 0.2, 100)  # Add noise to GIF
plt.step(x, y, label='Noisy Floor Function')
plt.grid(True)
plt.legend()
plt.show()
```
